```python
#Importing data analysis libraries

import pandas as pd
import numpy as np
from pandas import Series,DataFrame 

#importing visualization libraries
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Importing the train.csv file into a DataFrame called df

df= pd.read_csv('train.csv')
```

```python
#Displaying Raw dataFrame
df
```

```python
#Using Seaborns built-in command to make a catplot on columns sex and Pclass and visualizing it

sns.catplot('Sex',data=df,hue='Pclass',kind='count')
```

```python
#using Seanorns built-in command to make a catplot on columns sex and Pclass and visualizing it

sns.catplot('Pclass',data=df,hue='Sex', kind='count')
```

```python
#Creating a funtion called Male_femlale_children to asses the passenger as child if their age is less than 16

def Male_female_child(Passenger):
    age,sex= Passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
#Creating a mew column based on values in column "age" and "sex" and applying the fuction created previously on them

df['person']= df[['Age','Sex']].apply(Male_female_child,axis=1)
```

```python
#Displaying the first 10 rows

df[0:10]
```

```python
#Creating a new dataframe with no empty values

df_NoEmpty = df.dropna()
```

```python
#Changing datatype of column "Age" to int

df_NoEmpty = df_NoEmpty.astype({"Age":int})
```

```python
#Displaying the Dataframe

df_NoEmpty
```

```python
#Using seaborns built-in visualization command Facetgrid 
fig = sns.FacetGrid(df_NoEmpty,hue='Sex',aspect=4) #Specify datasource and column name for hue
fig.map(sns.kdeplot,'Age',shade=True) #Adding kedplot with facetgrid with map command
oldest = df_NoEmpty['Age'].max() # Creating max value for X axis to be max age in column
fig.set(xlim=(0,oldest)) # Max limit of Xaxis set to max age in column
fig.add_legend() # Adding legend to the figure
```

```python
#Using seaborns built-in visualization command Facetgrid
fig = sns.FacetGrid(df_NoEmpty,hue='Pclass',aspect=4) #Specify datasource and column name for hue
fig.map(sns.kdeplot,'Age',shade=True)  #Adding kedplot with facetgrid with map command
oldest = df_NoEmpty['Age'].max() # Creating max value for X axis to be max age in column
fig.set(xlim=(0,oldest)) # Max limit of Xaxis set to max age in column
fig.add_legend() # Adding legend to the figure
```

```python
#Creating New variable Deck with values of cabin colunm

Deck=df_NoEmpty['Cabin']
```

```python
levels = [] #Creating an empty array called levels
for level in Deck: #Creating a for loop that iterates values in Variable Deck
    levels.append(level[0]) #For each new value increase the count in array levels
Cabin_df= DataFrame(levels) # Converting array levels into a DataFrame
Cabin_df.columns=[ 'Cabin'] #Creatomg a mew column named Cabin
sns.catplot('Cabin',data=Cabin_df,kind='count') # Using Seaborn to viusalize the dataframe column Cabin
```

```python
Cabin_df=Cabin_df[Cabin_df.Cabin !='T'] #Deleting all values 'T' as it is not valid
```

```python
#Visualizing using Seaborns catplot method on column Cabin
sns.catplot('Cabin',data=Cabin_df,kind='count') 
```
